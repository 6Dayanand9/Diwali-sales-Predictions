# State

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values("Orders",ascending=False)
sales_state
```

#### ------------------------------------------------------------------------------------------------------------------------------
### give me only the 10 records we will use .head()

sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values("Orders",ascending=False).head(10)

sales_state
 #### -------------------------------------------------------------------------------------------------------------------------------

```python
fig=plt.figure(figsize=(20,9))
k=sns.barplot(data=sales_state,x="State",y='Orders',palette="bright")
plt.xlabel("STATE",fontsize=30)
plt.ylabel("ORDERS",fontsize=30,labelpad=50)
plt.title("STATE WISE ORDERS",fontsize=18)
k.set_xticklabels(k.get_xticklabels(),rotation=45,horizontalalignment="right",fontsize=15)
k.tick_params(axis="y",labelsize=15)

for bar in k.containers:
    k.bar_label(bar,fontsize=15)
plt.show()
```

#### from the above chart we found that the highest orders is from Uttar Pradesh and lowest from Telangana

## Total amount/sales from the state

```python
Amount_sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values("Amount",ascending=False)
#Amount_sales_state
```

```python
fig=plt.figure(figsize=(20,9))
k=sns.barplot(data=Amount_sales_state,x="State",y='Amount',palette="bright")
plt.xlabel("STATE",fontsize=30)
plt.ylabel("Amount",fontsize=30,labelpad=50)
plt.title("STATE WISE PURCHASING POWER",fontsize=18)
k.set_xticklabels(k.get_xticklabels(),rotation=45,horizontalalignment="right",fontsize=15)
k.tick_params(axis="y",labelsize=15)

"""for bar in k.containers:
    k.bar_label(bar,fontsize=15)"""
plt.show()
```

### Uttar Pradesh, Maharashtra, and Karnataka are the top three states in terms of both the highest number of orders and the greatest purchasing power.

#### Marital status Based on Gender

# Occupation

```python
fig=plt.figure(figsize=(10,5))
Oc=sns.countplot(data=df,x="Occupation")
Oc.set_xticklabels(Oc.get_xticklabels(),rotation=45,horizontalalignment="right")
for bar in Oc.containers:
    Oc.bar_label(bar)
plt.show()
```

```python
Sales_Occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
#Sales_Occupation

fig=plt.figure(figsize=(10,5))
k=sns.barplot(data=Sales_Occupation,x="Occupation",y="Amount")
k.set_xticklabels(k.get_xticklabels(),rotation=45,horizontalalignment="right")
plt.show()
```

#### from the above graph we can see that the most buyers are working in it,Healthcare and Aviation sector

# Product Category

```python
fig=plt.figure(figsize=(15,6))
pc=sns.countplot(data=df,x="Product_Category")
pc.set_xticklabels(pc.get_xticklabels(),rotation=45,horizontalalignment="right")
plt.xlabel("Product Category",fontsize=15)
pc.tick_params(axis="x",labelsize=12)
plt.show()
```

```python
product_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(7)
#product_sales
```

```python
fig=plt.figure(figsize=(9,4))
k=sns.barplot(data=product_sales,x="Product_Category",y="Amount")
k.set_xticklabels(k.get_xticklabels(),rotation=45,horizontalalignment="right")
plt.show()
```

### From the Above Graph we can see that the most sold products are from the Food ,Clothing and Electronics
